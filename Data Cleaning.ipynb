### Data Cleaning

```python
df["senior"] = df.job_title.apply(lambda x: 1 if "lead" in x.lower() or "senior" in x.lower() or "head" in x.lower() or
                               "principal" in x.lower() or "director" in x.lower() else 0)
df.senior.value_counts()
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop("salary", axis=1, inplace = True)
```

```python

df.drop("Unnamed: 0", axis=1, inplace = True)
```

```python
#regularize remote ratio
#parse Job title
#

#remote ratio
df["remote_ratio"] = df.remote_ratio/100
```

```python
#parse job title
df.job_title.value_counts()
```

```python
df["roles"] = df.job_title.apply(lambda x: "DS" if "data scien" in x.lower() else
                                "DE" if "data engine" in x.lower() or "etl" in x.lower() or "architect" in x.lower() else
                                "AN" if "analy" in x.lower() else
                                "ML" if "ml" in x.lower() or "machine" in x.lower() else
                                "RE" if "research" in x.lower() or "consult" in x.lower() else
                                "AI" if "ai" in x.lower() else
                                "CV" if "computer vision" in x.lower() else
                                "OT")
df.roles.value_counts()
```
