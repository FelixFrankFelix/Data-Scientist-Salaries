### Machine Learning Model building

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data_prepared.csv")
```

```python
df.head()
```

```python
## Building a base line Linear regression Model
X = df.drop("salary_in_usd",axis=1)
```

```python
X.columns.to_list()
```

```python
'''X = df[[ 'roles_AI',
 'roles_AN',
 'roles_CV',
 'roles_DE',
 'roles_DS',
 'roles_ML',
 'roles_OT',
 'roles_RE','senior','company_size','company_location','away','employment_type_FT']]'''

X = df["experience_level"]
```

```python
y = df.salary_in_usd.values
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
linmodel = LinearRegression()
```

```python
linmodel.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(linmodel,X_train,y_train,scoring = "r2", cv = 3)
scores
```

```python
scores.mean()
```

```python
y_pred = linmodel.predict(X_test)
```

```python
df = pd.DataFrame({"Actual":y_test,"Pred":y_pred})
df
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
treemodelbase = RandomForestRegressor(n_estimators = 20, random_state = 0)
```

```python
treemodelbase.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(treemodelbase,X_train,y_train,scoring = "r2", cv = 10)
scores
```

```python
scores.mean()
```

```python
y_pred = treemodelbase.predict(X_test)
df = pd.DataFrame({"Actual":y_test,"Pred":y_pred})
df
```

```python
r2_score(y_test,y_pred)
```

```python
#tuning Randomforestregressor
n_estimators = [int(x) for x in np.linspace(start = 5, stop=20, num = 1)]
max_features = ['sqrt','auto']
max_depth =[2,4]
min_samples_split =[2,4]
bootstrap = [True,False]
```

```python
param_grid = {'n_estimators':n_estimators,
              'max_features': max_features,
              'max_depth': max_depth,
              'min_samples_split':min_samples_split,
              'bootstrap':bootstrap,
    
}
print(param_grid)
```

```python
from sklearn.model_selection import GridSearchCV
treemodel = RandomForestRegressor()
```

```python
treemodel_Grid = GridSearchCV(estimator = treemodel, param_grid = param_grid, cv =3, verbose = 2, n_jobs = 4)
```

```python
treemodel_Grid.fit(X_train,y_train)
```

```python
treemodel_Grid.best_params_
```

```python
treemodel_Grid.best_estimator_.score(X_train,y_train)
```

```python
treemodel_Grid.best_estimator_.score(X_test,y_test)
```
