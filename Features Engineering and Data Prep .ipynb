# Features Engineering 

### Engineering Features
    :: Features Selection
    :: Features Encoding

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data_cleaned.csv")
```

```python
df.head()
```

#### Features Selection
   From Exploration, Features such as roles, seniority, company size, remote ratio, employment type and experience level are most importance in predicting the Data Scientist's salary while the work year and employee residence are also key.
    

```python
df.drop("job_title",axis=1,inplace = True)
```

```python
df.drop("salary_currency",axis=1,inplace = True)
df.columns
```

##### Features Encoding 
All of the features are Categorical 

```python
 df.head()
```

```python
dum = pd.get_dummies(df[["employment_type","roles"]])
```

```python
df = pd.concat([df,dum],axis=1)
```

```python
df
```

```python
df.drop(["employment_type","roles"],axis=1,inplace=True)
```

```python
df.work_year = df.work_year-2019
```

```python
df1 =df
df1.experience_level = df["experience_level"].apply(lambda x: 4 if x == "EX" else 3 if x == "SE" else 2 if x == "MI" else 1)
```

```python
df1.head()
```

```python
df1.columns
```

```python
df = df1
```

```python

df.company_size = df["company_size"].apply(lambda x: 3 if x == "L" else 2 if x == "M" else 1)
```

```python
df.head()
```

```python
df["away"]=0
```

```python
df.head()
```

```python
df.away[df.employee_residence!=df.company_location] = 1
```

```python
df.away.value_counts()
```

```python
df.drop(["employee_residence"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.company_location.value_counts()
```

```python
X_frequency_map = df.company_location.value_counts().to_dict()
df.company_location = df.company_location.map(X_frequency_map)
```
